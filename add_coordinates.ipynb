```python
import pandas as pd

data = pd.read_pickle('./data/task_1.pkl')


d = {'headline': [], 'city': [], 'country': [], 'lat': [], 'lon': [], 'countrycode': []}


for headline, city, country in data.values:
    d['headline'].append(headline)
    d['city'].append(city)
    d['country'].append(country)
    d['lat'].append(float(city['latitude']) if city else None)
    d['lon'].append(float(city['longitude']) if city else None)
    d['countrycode'].append(country['iso'] if country else None)

df = pd.DataFrame(d)
df.to_pickle('./data/task_2.pkl')
```
